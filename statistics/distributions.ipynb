```python
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Distribuciones continuas  <a class="anchor" id="continuas"></a>
### Exponencial <a class="anchor" id="exp"></a>
$
f_X(x;\lambda) = \begin{cases}
                     \lambda e^{-\lambda x} & x \geq 0 \\
                     0 & x<0
                 \end{cases}
$

```python
# par√°metro lambda
lam = 0.2

# definir variable aleatoria
x = st.expon(scale=1/lam, loc=0)

# valor del eje x para graficar
val = np.linspace(0, 40, 1000)

# figura que tendr√° los dos gr√°ficos
fig = plt.figure(figsize=(15, 5))

# gr√°fico de la distribuci√≥n (pdf)
ax = fig.add_subplot(121)
ax.set_title('pdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$f_X(x)$')
ax.plot(val, x.pdf(val))             # usar pdf para graficar

# gr√°fico de la distribuci√≥n (cdf)
ax = fig.add_subplot(122)
ax.set_title('cdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$F_X(x)$')
ax.plot(val, x.cdf(val))             # usar cdf para graficar
```

### Normal <a class="anchor" id="normal"></a>
$
f_X(x;\mu,\sigma^2) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{\left(x-\mu\right)^2}{2\sigma^2}}
$

```python
# par√°metro mu y sigma
mu = 10
sigma = 4

# definir variable aleatoria
x = st.norm(loc=mu, scale=sigma)

# valor del eje x para graficar
val = np.linspace(-5, 25, 1000)

# figura que tendr√° los dos gr√°ficos
fig = plt.figure(figsize=(15, 5))

# gr√°fico de la distribuci√≥n (pdf)
ax = fig.add_subplot(121)
ax.set_title('pdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$f_X(x)$')
ax.plot(val, x.pdf(val))             # usar pdf para graficar

# gr√°fico de la distribuci√≥n (cdf)
ax = fig.add_subplot(122)
ax.set_title('cdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$F_X(x)$')
ax.plot(val, x.cdf(val))             # usar cdf para graficar
```

### Uniforme <a class="anchor" id="uniforme"></a>

```python
# par√°metros a, b y c
a = 8
b = 22

# definir variable aleatoria
x = st.uniform(loc=a, scale=b-a)

# valor del eje x para graficar
val = np.linspace(0, 30, 1000)

# figura que tendr√° los dos gr√°ficos
fig = plt.figure(figsize=(15, 5))

# gr√°fico de la distribuci√≥n (pdf)
ax = fig.add_subplot(121)
ax.set_title('pdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$f_X(x)$')
ax.plot(val, x.pdf(val))             # usar pdf para graficar

# gr√°fico de la distribuci√≥n (cdf)
ax = fig.add_subplot(122)
ax.set_title('cdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$F_X(x)$')
ax.plot(val, x.cdf(val))             # usar cdf para graficar
```

### Triangular  <a class="anchor" id="tri"></a>

```python
# par√°metros a, b y c
a = 5
b = 25
c = 10

# definir variable aleatoria
x = st.triang((c-a)/(b-a), loc=a, scale=b-a)

# valor del eje x para graficar
val = np.linspace(0, 30, 1000)

# figura que tendr√° los dos gr√°ficos
fig = plt.figure(figsize=(15, 5))

# gr√°fico de la distribuci√≥n (pdf)
ax = fig.add_subplot(121)
ax.set_title('pdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$f_X(x)$')
ax.plot(val, x.pdf(val))             # usar pdf para graficar

# gr√°fico de la distribuci√≥n (cdf)
ax = fig.add_subplot(122)
ax.set_title('cdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$F_X(x)$')
ax.plot(val, x.cdf(val))             # usar cdf para graficar
```

### Trapezoidal  <a class="anchor" id="trap"></a>

```python
# par√°metros a, b y c
a = 2
b = 25
c = 6
d = 15

# definir variable aleatoria
x = st.trapz((c-a)/(b-a), (d-a)/(b-a), loc=a, scale=b-a)

# valor del eje x para graficar
val = np.linspace(0, 30, 1000)

# figura que tendr√° los dos gr√°ficos
fig = plt.figure(figsize=(15, 5))

# gr√°fico de la distribuci√≥n (pdf)
ax = fig.add_subplot(121)
ax.set_title('pdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$f_X(x)$')
ax.plot(val, x.pdf(val))             # usar pdf para graficar

# gr√°fico de la distribuci√≥n (cdf)
ax = fig.add_subplot(122)
ax.set_title('cdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$F_X(x)$')
ax.plot(val, x.cdf(val))             # usar cdf para graficar
```

## Distribuciones discretas  <a class="anchor" id="discretas"></a>
### Binomial <a class="anchor" id="binomial"></a>

```python
# par√°metros n y p
n = 10
p = 0.4

# definir variable aleatoria
x = st.binom(n, p)

# figura que tendr√° los dos gr√°ficos
fig = plt.figure(figsize=(15, 5))

# gr√°fico de la distribuci√≥n (pmf)
ax = fig.add_subplot(121)
ax.set_title('pmf')
ax.set_xlabel('$x$')
ax.set_ylabel('$p_X(x)$')
val = np.linspace(0, n, n+1)                          # valor del eje x para graficar
plt.plot(val, x.pmf(val), ls='None', marker='o')      # usar pmf para graficar

# gr√°fico de la distribuci√≥n (cdf)
ax = fig.add_subplot(122)
ax.set_title('cdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$F_X(x)$')
val = np.linspace(-1, n + 1, 1000)                    # valor del eje x para graficar
ax.plot(val, x.cdf(val))                              # usar cdf para graficar
```

### Poisson  <a class="anchor" id="poisson"></a>

```python
# par√°metros lambda
lam = 4

# definir variable aleatoria
x = st.poisson(lam)

# figura que tendr√° los dos gr√°ficos
fig = plt.figure(figsize=(15, 5))

# gr√°fico de la distribuci√≥n (pmf)
ax = fig.add_subplot(121)
ax.set_title('pmf')
ax.set_xlabel('$x$')
ax.set_ylabel('$p_X(x)$')
val = np.linspace(0, 14, 15)                          # valor del eje x para graficar
plt.plot(val, x.pmf(val), ls='None', marker='o')      # usar pmf para graficar

# gr√°fico de la distribuci√≥n (cdf)
ax = fig.add_subplot(122)
ax.set_title('cdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$F_X(x)$')
val = np.linspace(-1, 15, 1000)                       # valor del eje x para graficar
ax.plot(val, x.cdf(val))                              # usar cdf para graficar
```
